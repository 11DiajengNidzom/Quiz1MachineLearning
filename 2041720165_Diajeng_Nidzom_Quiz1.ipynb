1. Pastikan semua variabel kategorial diolah dengan baik. (Gunakan fitur mapping pada pandas)

```python
labels_sex = {
    'male': 1,
    'female': 0
}

labels_smoker ={
    'yes' : 1,
    'no' : 0
}

labels_region ={
    'northwest' : 0,
    'southwest' : 1,
    'southeast' : 2,
    'northeast' : 3 
}

#selanjutnya untuk mengencode sex,smoker,region pada labels
di['sex'] = di['sex'].map(labels_sex)
di['smoker'] = di['smoker'].map(labels_smoker)
di['region'] = di['region'].map(labels_region)

#selanjutnya adalah mengecek data
di.head()
```
